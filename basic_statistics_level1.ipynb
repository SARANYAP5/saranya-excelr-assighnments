# **Descriptive Analytics for Numerical Columns**

●	Objective:

 To compute and analyze basic statistical measures for numerical columns in the dataset.


●	Steps:


	Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	Identify numerical columns in the dataset.
	Calculate the mean, median, mode, and standard deviation for these columns.
	Provide a brief interpretation of these statistics.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
#read the file
sales = pd.read_csv('/content/sales_data_with_discounts.csv')
#nuemerical columns
numerical_columns=sales.select_dtypes(include=['int64','float64']).columns
#calculate statistics
results=[]
for c in numerical_columns:
  mean=np.mean(sales[c])
  median=np.median(sales[c])
  mode=stats.mode(sales[c])
  std=np.std(sales[c])
  results.append({'column':c,
                  'mean':mean,
                  'median':median,
                  'mode':mode,
                  'std':std})
#print result
for j in results:
  print(f"column:{j['column']}")
  print(f"mean:{j['mean']}")
  print(f"median:{j['median']}")
  print(f"mode:{j['mode']}")
  print(f"std:{j['std']}\n")
```

```python
sales=pd.read_csv('/content/sales_data_with_discounts.csv')
numerical_columns=sales.select_dtypes(include=['int64','float64']).columns
for c in numerical_columns:
  plt.hist(sales[c])
  plt.title(f"histogram{c}")
  plt.xlabel(c)
  plt.ylabel('frequency')
  plt.show()
  #analyze the distribution of categories
  skewness=sales[c].skew()
  if(skewness>1):
    skew_label='more skew right'
  elif(skewness<-1):
    skew_label='more skew left'
  elif(skewness>0.5):
    skew_label='moderate skew right'
  elif(skewness<-0.5):
    skew_label='moderate skew left'
  else:
    skew_label='symmetry'
  Q1=sales[c].quantile(0.25)
  Q3=sales[c].quantile(.75)
  IQR=Q3-Q1
  upper_bond=Q3+1.5*IQR
  lower_bond=Q1-1.5*IQR
  outliers=sales[(sales[c]<lower_bond)|(sales[c]>upper_bond)]
  print(f"column name{c}")
  print(f"{skewness}{skew_label}")
  print(f"no of outliers:{outliers.shape[0]}")
  if outliers.shape[0]>0:
    print(f"outliers:{c}{outliers}")

```

Standardization of Numerical Variables


●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

●	Steps:

	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma

	Show before and after comparisons of the data distributions.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sales=pd.read_csv('/content/sales_data_with_discounts.csv')
numerical_columns=sales.select_dtypes(include=['int64','float64']).columns
for c in numerical_columns:
  mean=np.mean(sales[c])
  std=np.std(sales[c])
  z_sales[c]=(sales[c]- mean)/std
  print(f"column name:{c}")

  # Example comparing distributions before and after standardization
  plt.figure(figsize=(12, 6))

  # Plot before standardization
  plt.subplot(1, 2, 1)
  plt.title('Before Standardization')
  plt.hist(sales[c], color='blue')
  plt.xlabel('Value')
  plt.ylabel('Frequency')

  # Plot after standardization
  plt.subplot(1, 2, 2)
  plt.title('After Standardization')
  plt.hist(z_sales[c],bins=20,color='blue')
  plt.xlabel('Value')
  plt.ylabel('Frequency')
  plt.show()
```
